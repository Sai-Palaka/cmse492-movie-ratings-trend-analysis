```python
# load dataset using pandas and explore it
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.DataFrame({
    "year": [2000, 2005, 2010, 2015, 2020],
    "rating": [6.5, 7.0, 7.5, 7.2, 7.8]
})

# Display first few rows
print("First few rows of the dataset:")
print(df.head())

# Display dataset information
print("\nDataset information:")
print(df.info())

# Display summary statistics
print("\nSummary statistics:")
print(df.describe())

# Calculate average rating by year
avg_ratings = df.groupby('year')['rating'].mean()

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(avg_ratings.index, avg_ratings.values, marker='o', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Movie Ratings Over Time')
plt.grid(True, alpha=0.3)
plt.show()
```
